```python
from awpy import Demo

def get_bomb_carried_for_stat(dem: Demo, round_num, player):
    # Retrieve the start and end ticks for the round
    round_data = dem.rounds[dem.rounds['round'] == round_num]
    start = round_data['freeze_end'].values[0]
    end = round_data['end'].values[0]
    
    # Calculate round length
    round_len = end - start
    
    # Filter ticks for the relevant player and round duration
    player_ticks = dem.ticks[(dem.ticks['tick'] >= start) & 
                             (dem.ticks['tick'] <= end) & 
                             (dem.ticks['name'] == player)]
    
    # Check if 'C4' is in the inventory column across all filtered ticks
    bomb_carriage_ticks = player_ticks['inventory'].apply(lambda inv: 'C4' in inv)
    
    # Count how many ticks the bomb is carried
    tick_counter_for_bomb_carriage = bomb_carriage_ticks.sum()
    
    # Return the proportion of time the bomb was carried during the round
    return tick_counter_for_bomb_carriage / round_len

def get_awp_used_in_round(dem: Demo, round_num, player):
    # Retrieve the start and end ticks for the round
    round_data = dem.rounds[dem.rounds['round'] == round_num]
    start = round_data['freeze_end'].values[0]
    end = round_data['end'].values[0]
    
    # Calculate round length
    round_len = end - start
    
    # Filter ticks for the relevant player and round duration
    player_ticks = dem.ticks[(dem.ticks['tick'] >= start) & 
                             (dem.ticks['tick'] <= end) & 
                             (dem.ticks['name'] == player)]
    
    # Check if 'C4' is in the inventory column across all filtered ticks
    awp_carried_ticks = player_ticks['inventory'].apply(lambda inv: 'AWP' in inv)

    return awp_carried_ticks.sum() > 0

def get_distance_travelled_stat(dem: Demo, round_num, player):
    import numpy as np
    # Retrieve the start and end ticks for the round
    round_data = dem.rounds[dem.rounds['round'] == round_num]
    start = round_data['freeze_end'].values[0]
    end = round_data['end'].values[0]
    
    # Filter ticks for the relevant player and round duration
    player_ticks = dem.ticks[(dem.ticks['tick'] >= start) & 
                             (dem.ticks['tick'] <= end) & 
                             (dem.ticks['name'] == player)]
    
    # Sort the ticks by time to calculate distance between consecutive positions
    player_ticks = player_ticks.sort_values('tick')
    
    # Calculate the Euclidean distance between consecutive ticks
    distances = np.sqrt(
        np.diff(player_ticks['X'])**2 +
        np.diff(player_ticks['Y'])**2 +
        np.diff(player_ticks['Z'])**2
    )
    
    # Sum the distances to get the total distance traveled
    total_distance_traveled = distances.sum()
    
    return total_distance_traveled

def get_last_alive_flag(dem: Demo, player, round_num, name_team_dict):
    last_alive_tick_for_player = dem.ticks[(dem.ticks['round']==round_num)&(dem.ticks['name']==player)&(dem.ticks['health']>0)].sort_values('tick', ascending=False)['tick'].values[0]
    alive_teammates_last_alive_tick = dem.ticks[(dem.ticks['team_clan_name']==name_team_dict[player])&(dem.ticks['name']!=player)&(dem.ticks['health']>0)&(dem.ticks['tick']==last_alive_tick_for_player)]
    return len(alive_teammates_last_alive_tick) == 0

def get_team_equipment_value(dem: Demo,round_num, player, name_team_dict):
    round_data = dem.rounds[dem.rounds['round'] == round_num]
    start = round_data['freeze_end'].values[0]

    return dem.ticks[(dem.ticks['tick'] == start)&(dem.ticks['team_clan_name'] == name_team_dict[player])]['current_equip_value'].sum()

def get_enemy_equipment_value(dem: Demo,round_num, player, name_team_dict):
    round_data = dem.rounds[dem.rounds['round'] == round_num]
    start = round_data['freeze_end'].values[0]

    return dem.ticks[(dem.ticks['tick'] == start)&(dem.ticks['team_clan_name'] != name_team_dict[player])]['current_equip_value'].sum()

def create_df_for_cluster(dem: Demo):
    import pandas as pd
    from tqdm import tqdm
    
    name_team_dict = {item.split('_')[0]: item.split('_')[1] for item in set(dem.ticks['name'] +'_'+ dem.ticks['team_clan_name'])}
    result_df = pd.DataFrame(columns=[
        'name', 'team_name', 'team_side', 'team_equipment_value', 'enemy_equipment_value', 'match_name', 'round_num',
        'kills_in_round', 'deaths_in_round', 'dmg_in_round', 'dmg_taken_in_round', 'dmg_in_round_util',
        'entry_kill', 'entry_damages', 'round_survived', 'bomb_carried_ratio', 'distance_travelled',
        'attacking_util_threw', 'supporting_util_threw', 'last_alive_flag', 'hs_kill_count',
        'kill_count', 'kill_count_team', 'kill_participate_ratio', 'assist_count', 
        'assist_participate_ratio', 'awp_used_flag', 'engagement_count'
    ])
    round_count = dem.rounds['round'].max()

    error_count = 0

    for round_num in (range(1, round_count + 1)):
        # rounds_to_exclude = [1, 13] have to exclude rounds in later phase
        rounds_to_exclude = []
        if round_num in rounds_to_exclude:
            continue
        for player in name_team_dict.keys():
            try:
                # print(f"team: {name_team_dict[player]}, player: {player}, round: {round_num}")
                team_side = dem.ticks[(dem.ticks['name']==player)&(dem.ticks['round'] == round_num)]['team_name'].values[0]
                team_name = name_team_dict[player]
                team_equipment_value = get_team_equipment_value(dem,round_num,player,name_team_dict)
                enemy_equipment_value = get_enemy_equipment_value(dem, round_num, player, name_team_dict)
                match_name = str(dem.path)
                kills_in_round = len(dem.kills[(dem.kills['round'] == round_num) & (dem.kills['attacker_name'] == player)])
                deaths_in_round = len(dem.kills[(dem.kills['round'] == round_num) & (dem.kills['victim_name'] == player)])
                dmg_in_round = dem.damages[(dem.damages['round'] == round_num) & (dem.damages['attacker_name'] == player)]['dmg_health_real'].sum()
                dmg_taken_in_round = dem.damages[(dem.damages['round'] == round_num) & (dem.damages['victim_name'] == player)]['dmg_health_real'].sum()
                dmg_in_round_util = dem.damages[(dem.damages['round'] == round_num) & (dem.damages['attacker_name'] == player) & (dem.damages['weapon'].isin(['inferno', 'hegrenade']))]['dmg_health_real'].sum()
                entry_kill = player == dem.kills[dem.kills['round'] == round_num].sort_values('tick', ascending = True)['attacker_name'].values[0] if len(dem.kills[dem.kills['round'] == round_num]) > 0 else False
                entry_damages = player in dem.damages[dem.damages['round'] == round_num].sort_values('tick', ascending = True)['attacker_name'].values[:10] if len(dem.damages[dem.damages['round'] == round_num]) > 0 else False
                round_survived = 1 - deaths_in_round
                bomb_carried_ratio = get_bomb_carried_for_stat(dem,round_num,player)
                distance_travelled = get_distance_travelled_stat(dem,round_num,player)
                attacking_util_threw = len(dem.grenades[(dem.grenades['round']==round_num) & (dem.grenades['thrower'] == player) & (dem.grenades['grenade_type'].isin(['molotov','he_grenade','incendiary_grenade']))].drop_duplicates(subset = ['entity_id']))
                supporting_util_threw = len(dem.grenades[(dem.grenades['round']==round_num) & (dem.grenades['thrower'] == player) & (dem.grenades['grenade_type'].isin(['smoke','flashbang']))].drop_duplicates(subset = ['entity_id']))
                last_alive_flag = get_last_alive_flag(dem,player,round_num,name_team_dict)
                hs_kill_count = len(dem.kills[(dem.kills['attacker_name']==player)&(dem.kills['headshot'])&(dem.kills['round']==round_num)])
                kill_count = len(dem.kills[(dem.kills['attacker_name']==player)&(dem.kills['round']==round_num)])
                kill_count_team = len(dem.kills[(dem.kills['attacker_team_clan_name']==name_team_dict[player])&(dem.kills['round']==round_num)])
                kill_participate_ratio = kill_count / kill_count_team if kill_count_team != 0 else 0
                assist_count = len(dem.kills[(dem.kills['assister_name']==player)&(dem.kills['round']==round_num)])
                assist_participate_ratio = assist_count / kill_count_team if kill_count_team != 0 else 0
                awp_used_flag = get_awp_used_in_round(dem,round_num,player)
                engagement_count = len(dem.damages[(dem.damages['round']==round_num)&((dem.damages['attacker_name']==player)|(dem.damages['victim_name']==player))])

                temp_df = pd.DataFrame([{
                    'name': player, 'team_name': team_name, 'team_side': team_side, 
                    'team_equipment_value': team_equipment_value, 'enemy_equipment_value': enemy_equipment_value, 
                    'match_name': match_name, 'round_num': round_num, 'kills_in_round': kills_in_round, 'deaths_in_round': deaths_in_round, 
                    'dmg_in_round': dmg_in_round, 'dmg_taken_in_round': dmg_taken_in_round, 
                    'dmg_in_round_util': dmg_in_round_util, 'entry_kill': entry_kill, 
                    'entry_damages': entry_damages, 'round_survived': round_survived, 
                    'bomb_carried_ratio': bomb_carried_ratio, 'distance_travelled': distance_travelled, 
                    'attacking_util_threw': attacking_util_threw, 'supporting_util_threw': supporting_util_threw, 
                    'last_alive_flag': last_alive_flag, 'hs_kill_count': hs_kill_count, 
                    'kill_count': kill_count, 'kill_count_team': kill_count_team, 
                    'kill_participate_ratio': kill_participate_ratio, 'assist_count': assist_count, 
                    'assist_participate_ratio': assist_participate_ratio, 'awp_used_flag': awp_used_flag, 
                    'engagement_count': engagement_count
                }])
                result_df = pd.concat([result_df, temp_df], ignore_index=True)
            except:
                error_count += 1

    return result_df, error_count

```

```python
from awpy.plot import gif, PLOT_SETTINGS
from tqdm import tqdm

frames = []

for tick in tqdm(dem.ticks[dem.ticks["round"] == 1].tick.values[::64]):
    frame_df = dem.ticks[dem.ticks["tick"] == tick]
    frame_df = frame_df[
        ["X", "Y", "Z", "health", "armor_value", "pitch", "yaw", "team_name", "name"]
    ]

    points = []
    point_settings = []

    for _, row in frame_df.iterrows():
        points.append((row["X"], row["Y"], row["Z"]))

        # Determine team and corresponding settings
        team = "ct" if row["team_name"] == "CT" else "t"
        settings = PLOT_SETTINGS[team].copy()

        # Add additional settings
        settings.update(
            {
                "hp": row["health"],
                "armor": row["armor_value"],
                "direction": (row["pitch"], row["yaw"]),
                "label": row["name"],
            }
        )

        point_settings.append(settings)

    frames.append({"points": points, "point_settings": point_settings})

print("Finished processing frames. Creating gif...")
gif("de_ancient", frames, "de_dust2.gif", duration=100)
```

### Distribution of Affine Transformation of Multivariate Normal

If $x_1$ and $x_2$ are i.i.d. standard normal distributed random variables, then the vector $X = \begin{bmatrix} x_1 \\ x_2 \end{bmatrix}$ follows a multivariate normal distribution with mean vector $\mathbf{0}$ and covariance matrix $\mathbf{I}$, denoted as $X \sim N(\mathbf{0}, \mathbf{I})$.

Given a constant vector $\mu$ and a constant matrix $\mathbf{B}$, the affine transformation $\mu + \mathbf{B}X$ also follows a multivariate normal distribution. 

**The distribution of** $\mu + \mathbf{B}X$ **is:**
 $$  \mu + \mathbf{B}X \sim N(\mu + \mathbf{B}\mathbf{0}, \mathbf{B}\mathbf{I}\mathbf{B}^T) = N(\mu, \mathbf{B}\mathbf{B}^T). $$

**Therefore, with** $\mu = \begin{bmatrix} 1.5 \\ -0.2 \end{bmatrix}^T$ **and** $\mathbf{B} = \begin{bmatrix} 0.3 & -0.6 \\ 2.0 & 3.1 \end{bmatrix}$, **the distribution of** $\mu + \mathbf{B}X$ **is:**

$$ \mu + \mathbf{B}X \sim N \left( \begin{bmatrix} 1.5 \\ -0.2 \end{bmatrix}, \begin{bmatrix} 0.3 & -0.6 \\ 2.0 & 3.1 \end{bmatrix} \begin{bmatrix} 0.3 & 2.0 \\ -0.6 & 3.1 \end{bmatrix} \right) = N \left( \begin{bmatrix} 1.5 \\ -0.2 \end{bmatrix}, \begin{bmatrix} 0.45 & -1.26 \\ -1.26 & 13.61 \end{bmatrix} \right). $$

**Explanation:**

* **Affine Transformation:** An affine transformation is a linear transformation followed by a translation. In this case,  $\mathbf{B}X$ is the linear transformation and adding  $\mu$ is the translation.
* **Multivariate Normal Distribution:**  A multivariate normal distribution is a generalization of the one-dimensional normal distribution to multiple dimensions. It is characterized by its mean vector and covariance matrix.
* **Mean and Covariance of Affine Transformation:** The mean of an affine transformation of a random variable is simply the affine transformation of the original mean. The covariance matrix of an affine transformation is obtained by multiplying the original covariance matrix by the transformation matrix and its transpose.

This concept is important in many areas of statistics and probability, including portfolio optimization and risk management. For example, in portfolio optimization, the returns of different assets are often modeled as multivariate normal random variables. An affine transformation can then be used to model the return of a portfolio of assets.
